```python
def parser(st):
  return '['+st[1:-1].replace("'",'"').replace('True','true').replace('False','false').replace('generate_report','generate_work_report')+']'
```

```python
from json.decoder import JSONDecodeError
err_id = []
data_final = pd.DataFrame(columns=['Query','tool_ids','pos_args','negative_ids','output'])

def generate(df):
  tool_null =[]
  for i in range(len(df)):
    try:
      ls = json.loads(parser(df['output'].loc[i]))
      lst = [x['tool_name'] for x in ls]
      try:
        lss = [tool_dct[x] for x in lst]
        a = range(len(samyak_tools_list))
        c = [x for x in a if x not in lss]
        pos_arg = []
        for x in ls:
          y = [z['argument_name'] for z in x['arguments']]
          pos_arg.append(y)
          #print(x['arguments'])
          neg_id = random.choice(c)
        #print(c)
        data_final.loc[len(data_final)] =[df['query'].loc[i],lss,pos_arg,neg_id,df['output'].loc[i]]
        #print('hii')
      except KeyError:
        #print(f"didnt found {lss}")
        err_id.append(i)

    except JSONDecodeError:
      print(f"error in {i}:-")
      print(parser(df['output'].loc[i]))
      #err_id.append(i)
      tool_null.append(i)
  print(tool_null)
  print(err_id)
```

```python
generate(pd.read_csv('samyakfinaldata2.csv'))
generate(pd.read_csv('samyakfinaldata3.csv'))
generate(pd.read_csv('samyakfinaldata4.csv'))
generate(pd.read_csv('samyakfinaldata5.csv'))
```

```python
tool_argument_dct = {}
for tool in samyak_tools_list:
  js = json.loads(tool)
  tool_name = js['tool_name']
  #print(tool_name)
  args = dict()
  for arg in js['arguments']:
    args[arg['name']]= {'Tool':tool_name,'Tool Description':js['tool_description'],'Tool ArgumentName':arg['name'],'Tool Argument Description':arg['description'],'type':arg['type']}
  tool_argument_dct[tool_name]=args
```

```python
data_list = []
for i in range(len(data_final)):
  pst_ctx = []
  for id_arg in range(len(data_final['pos_args'].loc[i])):
    arg_list = data_final['pos_args'].loc[i][id_arg]
    tool_name = json.loads(samyak_tools_list[data_final['tool_ids'].loc[i][id_arg]])['tool_name']
    for argument in arg_list:
      try:
        dic = {'title':tool_name+' '+argument, 'text': str(tool_argument_dct[tool_name][argument]), 'score':10000,'title_score':1,'passage_id':12345}
        pst_ctx.append(dic)
      except KeyError:
        pass
  neg_ctx = []
  for id_arg in range(len(data_final['pos_args'].loc[i])):
    arg_list = data_final['pos_args'].loc[i][id_arg]
    tool_name = json.loads(samyak_tools_list[data_final['tool_ids'].loc[i][id_arg]])['tool_name']
    js = [arg['name'] for arg in json.loads(samyak_tools_list[tool_dct[tool_name]])['arguments']]
    for argument in js:
      if argument not in arg_list:
        try:
          dic = {'title':tool_name+' '+argument, 'text': str(tool_argument_dct[tool_name][argument]), 'score':1000,'title_score':0,'passage_id':1234}
          neg_ctx.append(dic)
        except KeyError:
          pass
  hard_neg_ctx=[]
  neg_tool_name = json.loads(samyak_tools_list[data_final['negative_ids'].loc[i]])['tool_name']
  for args in tool_argument_dct[neg_tool_name]:
    hard_neg_ctx.append({'title':neg_tool_name+' '+args,'text':str(tool_argument_dct[neg_tool_name][args]),'score':10,'title_score':-1,'passage_id':1234})
  data_point ={
        "dataset": "finetune",
        "question": data_final['Query'].loc[i],
        "answers": [str(data_final['output'].loc[i])], #[tool_dct[x] for x in data_final['tool_ids'].loc[i]],
        #"positive_ctxs": list of dictionaries of format {'title': str, 'text': str, 'score': int, 'title_score': int, 'passage_id': str}
        #"negative_ctxs": []
        #"hard_negative_ctxs": list of dictionaries of format {'title': str, 'text': str, 'score': int, 'title_score': int, 'passage_id': str}
        "positive_ctxs":pst_ctx,
        "negative_ctxs":[], #neg_ctx,
        "hard_negative_ctxs":hard_neg_ctx
    }
  data_list.append(data_point)
  #print('yay')
```
